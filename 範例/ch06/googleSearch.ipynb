```python
dialogue = []      # 對話記錄串列
def get_response(dialogue, streamTF=False):
  response = openai.chat.completions.create(
      model = 'gpt-3.5-turbo',
      messages = dialogue,
      stream = streamTF                   # 設定流式傳輸生成回覆
  )
  if streamTF:
    for chunk in response:              # response 為可迭代物件
      if chunk.choices[0].delta.content is not None:
        yield chunk.choices[0].delta.content  # 傳出片段回覆文本
  else:
    yield response.choices[0].message.content # 傳出所有回覆文本
```

```python
def get_webRes(query):
  googleAns = '以下是google搜尋的結果:\n'
  for item in search(query, num_results=4, advanced=True):
    googleAns += f'標題：{item.title}\n'
    googleAns += f'描述：{item.description}\n\n'
  return googleAns
```

```python
saveTalk = 2       # 保留最後2組提問和回覆記錄
def chat(query, streamTF, googleTF=False):
  global dialogue            # 宣告dialogue屬於全域變數
  if googleTF:
    sysMsg = get_webRes(query)    # 網路搜尋內容
    dialogue.append({'role':'system','content':f'請依"{sysMsg}"提示使用繁體中文回覆。'})
    dialogue.append({'role':'user','content':query})     # 存放提問資料
  else:
    dialogue.append({'role':'user','content':query})     # 提問文句不含網路搜尋內容

  ans_all = ''
  for answer in get_response(dialogue, streamTF):  # 呼叫函式生成片段文本
    ans_all += answer
    yield answer

  dialogue.append({'role':'assistant','content':ans_all})  # 新增元素
  while len(dialogue) > saveTalk * 2:  # 當提問和回覆記錄超過2組時
    dialogue.pop(0)                   # 把前面的串列元素移除掉
```

```python
def gather_chat(query, streamTF, googleTF):
  sentence = ''
  for ans in chat(query, streamTF, googleTF):
    sentence += ans             # 累加片段文句
    yield sentence
```

```python
txtAsk = gr.Textbox(label='輸入您的問題：')
chkStream = gr.Checkbox(label='流式傳輸', value=False)
chkGoogle = gr.Checkbox(label='google搜尋', value=False)
gr.Interface(
    fn = gather_chat,
    inputs = [txtAsk, chkStream, chkGoogle],
    outputs = gr.Textbox(label='AI 的回答：'),
    title = 'AI 聊天室'
).queue().launch(share=True)
```
