```python
def get_webRes(query):
  cx = '搜尋引擎 ID'               # 您的搜尋引擎 ID
  key = 'Custom Search JSON API 金鑰'    # 您的 API 金鑰
  q = query                              # 要搜尋的文字
  num = 2                                # 傳回的搜尋結果數量
  url = 'https://www.googleapis.com/customsearch/v1' # 發送請求的網址
  params = {'cx':cx, 'key':key, 'q':q, 'num':num,'lr':'lang_zh-TW'}   # 查詢參數
  reply = requests.get(url, params=params)   # 傳回JSON格式的資料
  searchAns = '以下是搜尋引擎查詢的結果:\n'
  for item in reply.json()['items']:
    searchAns += f'標題：{item["title"]}\n'
    searchAns += f'描述：{item["snippet"]}\n\n'
    return searchAns
```

```python
def chat(query):
  sysMsg = get_webRes(query)    # 網路搜尋內容
  response = openai.chat.completions.create(
      model = 'gpt-3.5-turbo',
      messages = [
           {'role':'system', 'content':f'請依"{sysMsg}"提示回覆。'},
           {'role': 'user', 'content': query}
      ]
  )
  return response.choices[0].message.content
```

```python
gr.Interface(
    fn = chat,
    inputs = gr.Textbox(label='輸入您的問題：'),
    outputs= gr.Textbox(label='AI 的回答：'),
    title = 'AI 聊天室'
).queue().launch(share=True)
```
