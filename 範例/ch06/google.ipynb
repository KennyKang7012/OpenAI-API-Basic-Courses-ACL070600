```python
def get_webRes(query):
  googleAns = '以下是google搜尋的結果:\n'
  for item in search(query, num_results=4, advanced=True):
    googleAns += f'標題：{item.title}\n'
    googleAns += f'描述：{item.description}\n\n'
  return googleAns
```

```python
def chat(query):
  sysMsg = get_webRes(query)    # 網路搜尋內容
  response = openai.chat.completions.create(
      model = 'gpt-3.5-turbo',
      messages = [
          {'role':'system', 'content':f'請依"{sysMsg}"提示使用繁體中文回覆。'},
          {'role': 'user', 'content': query}
      ]
  )
  return response.choices[0].message.content
```

```python
gr.Interface(
    fn = chat,
    inputs = gr.Textbox(label='輸入您的問題：'),
    outputs = gr.Textbox(label='AI 的回答：'),
    title = 'AI 聊天室'
).queue().launch(share=True)
```
