```python
def Paint(ask,num):
  openai.api_key = 'OpenAIAPI金鑰'
  response=openai.images.generate(
    model='dall-e-2',
    prompt=ask,
    size='512x512',
    n=num
  )
  image_urls=[]
  i=0
  while(i<len(response.data)):
    image_urls.append(response.data[i].url)
    i+=1
  return image_urls
```

```python
gr.Interface(
  fn=Paint,
  inputs= [gr.Textbox(label='請輸入圖片的描述：'),
       gr.Slider(minimum=1,maximum=4,step=1,value=2,label='張數')],
  outputs= 'gallery'
).queue().launch()
```
