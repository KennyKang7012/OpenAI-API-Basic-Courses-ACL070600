```python
def Paint(img_s,img_m,ask):
  openai.api_key='OpenAIAPI金鑰'
  response=openai.images.edit(
    image=open('source.png','rb'),
    mask=open('mask.png','rb'),
    prompt=ask,
    size='1024x1024'
  )
  return response.data[0].url
```

```python
gr.Interface(
  fn=Paint,
  inputs=[gr.Image(label='來源圖片：',value='source.png',height=128,width=128),
      gr.Image(label='遮罩圖片：',value='mask.png',height=128,width=128),
      gr.Textbox(label='請輸入插入圖形的描述：')],
  outputs='image'
).queue().launch()
```
