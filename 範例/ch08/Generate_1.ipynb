```python
def Paint(ask):
  openai.api_key = 'OpenAIAPI金鑰'
  response=openai.images.generate(
    model='dall-e-3',
    prompt=ask,
    size='1024x1024'
  )
  return response.data[0].url
```

```python
gr.Interface(
  fn=Paint,
  inputs= gr.Textbox(label='請描述想畫圖形的內容：'),
  outputs= gr.Image(label='AI 的畫作：')
).queue().launch()
```
