```python
def Paint(ask,u_model,u_size,u_n,u_quality,u_style):
  openai.api_key = 'OpenAIAPI金鑰'
  if u_model=='dall-e-3':
    u_n=1
    if u_size=='A':
      u_size='1024x1024'
    elif u_size=='B':
      u_size='1792x1024'
    else:
      u_size='1024x1792'
  else:
    if u_size=='A':
      u_size='1024x1024'
    elif u_size=='B':
      u_size='512x512'
    else:
      u_size='256x256'
  response=openai.images.generate(
    model=u_model,
    prompt=ask,
    size=u_size,
    n=u_n,
    quality=u_quality,
    style=u_style
  )
  image_urls=[]
  i=0
  while(i<len(response.data)):
    image_urls.append(response.data[i].url)
    i+=1
  return image_urls
```

```python
gr.Interface(
  fn=Paint,
  inputs=[gr.Textbox(label='請描述想畫圖形的內容：'),
    gr.Radio(label='模型：',choices=['dall-e-3','dall-e-2'],value='dall-e-3'),
    gr.Radio(label='尺寸：',choices=['A','B','C'],value='A'),
    gr.Slider(label='張數：',minimum=1,maximum=10,step=1,value=1),
    gr.Radio(label='品質：',choices=['standard','hd'],value='standard'),
    gr.Radio(label='風格：',choices=['vivid','natural'],value='vivid')],
  outputs='gallery'
).queue().launch()
```
