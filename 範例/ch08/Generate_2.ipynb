```python
def Paint(com,kind,style):
  openai.api_key = 'OpenAIAPI金鑰'
  response=openai.images.generate(
    model='dall-e-3',
    prompt=f'以{style}風格設計{kind}類型公司的商標,文字為:「{com}」'
  )
  return response.data[0].url
```

```python
gr.Interface(
  fn=Paint,
  inputs= [gr.Textbox(label='公司名稱：'),
      gr.Textbox(label='營業項目：'),
      gr.Dropdown(choices=['活潑','正式','抽象'],label='風格：')],
  outputs= gr.Image(label='公司商標：'),
  title='商標生成器'
).queue().launch()
```
