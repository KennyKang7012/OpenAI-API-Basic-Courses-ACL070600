```python
def fnChat(prompt):
   with open('飯店服務資訊.txt', 'r', encoding='utf-8') as file:
     hotel_info = file.read()
   openai.api_key = 'OpenAIAPI金鑰'   # 金鑰
   response = openai.chat.completions.create(
       model = 'gpt-3.5-turbo',
       stream = True,
       messages = [
        {'role':'system', 'content':f'你是飯店客服人員, 請依"{hotel_info}"提示回覆，並用繁體中文回覆。'},
        {'role': 'user', 'content': prompt}
    ]
   )
   for chunk in response:           # response 為可迭代物件
     if chunk.choices[0].delta.content is not None:
       yield chunk.choices[0].delta.content # 傳出片段回覆文本
```

```python
def fnHotelChatServices(ask):
  sentence = ''
  for ans in fnChat(ask):
    sentence += ans             # 累加片段文句
    yield sentence
```

```python
txtAsk = gr.Textbox(label='輸入您的問題：')
gr.Interface(
    fn = fnHotelChatServices,
    inputs = txtAsk,
    outputs= gr.Textbox(label='碁峰客服機器人的回答：'),
    title = '碁峰飯店智慧客服'
).queue().launch(share=True)
```
