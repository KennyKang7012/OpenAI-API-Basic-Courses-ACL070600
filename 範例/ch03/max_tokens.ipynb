```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'user', 'content': 'OpenAI 你好!'}
	],
    max_tokens = 10                # 設定回覆訊息時,最多可使用10個token
)
```

```python
print(response.choices[0].message.content)      # 輸出回覆訊息內容
```

```python
print(response.choices[0].finish_reason)        # 輸出結束回覆的原因
print(response.usage.prompt_tokens)             # 輸出提問時所使用的token數量
print(response.usage.completion_tokens)         # 輸出回覆時所使用的token數量
print(response.usage.total_tokens)              # 輸出問答所使用的總token數量
```
