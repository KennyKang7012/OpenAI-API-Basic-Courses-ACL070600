```python
encoder = tiktoken.encoding_for_model('gpt-3.5-turbo')
```

```python
print(encoder.name)            # 顯示編碼器名稱
```

```python
token = encoder.encode('OpenAI is good !')   # 將文本轉換成token串列
print(token)
```

```python
text = encoder.decode([5109,15836,374,1695,758])  #將token串列轉換成文本
print(text)
```
