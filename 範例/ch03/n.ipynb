```python
!pip install openai
import openai
openai.api_key = 'OpenAIAPI金鑰'  # 金鑰
```

```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'user', 'content': '請指出彩虹的顏色'}
	],
    n = 2                  # 設定回覆生成的訊息數量
)
```

```python
print(response.choices[0].message.content)      # 輸出第一筆回覆訊息的文本
print(response.choices[1].message.content)      # 輸出第二筆回覆訊息的文本
```

```python
for reply in response.choices:                    # 依序取出回覆choices串列內的元素
    print (reply.index, reply.message.content)    # 輸出取出元素的回覆文本
```
