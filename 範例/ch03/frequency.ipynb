```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [{'role': 'user', 'content': '請用40個字介紹北極熊'}],
    frequency_penalty = -2            # 參數值設很低,使語句重複出現概率變大
)
```
