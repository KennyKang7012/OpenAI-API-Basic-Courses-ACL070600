```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'user', 'content': '列出10個台灣小吃名稱'}
	],
    stop = ['血','肉']             # 設定使回覆生成中斷的字串串列
)
```

```python
print(response.choices[0].message.content)    # 輸出回覆訊息的文本
```
