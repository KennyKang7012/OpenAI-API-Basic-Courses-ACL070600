```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'user', 'content': '僅列出一項台灣小吃名稱'}
	],
    n=2,                        # 設定回覆生成的訊息數量
    temperature = 0.1           # 參數值設很低,使回覆訊息重複性高
)
```

```python
print(response.choices[0].message.content)    # 輸出第一筆回覆訊息的文本
print(response.choices[1].message.content)    # 輸出第二筆回覆訊息的文本
```
