```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'user', 'content': '請用35個以內的字介紹台北巿'}
	],
    n = 2,                 # 設定回覆生成的訊息數量
    top_p = 0.25           # 參數值設很低,使備選語句集合元素少
)
```

```python
for reply in response.choices:
    print (reply.index, reply.message.content)
```
