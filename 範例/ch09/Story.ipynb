```python
import base64
import requests
```

```python
def EncodeImage(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
def Vision(img1,img2,style,num):
  api_key = "OpenAIAPI金鑰"
  base64_image_1 = EncodeImage(img1)
  base64_image_2 = EncodeImage(img2)
  headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
  }
  payload = {
    "model":"gpt-4o",
    "messages":[
      {
        "role":"user",
        "content": [
          {"type":"text","text":f"使用繁體中文根據兩張圖像創作約{num}字數{style}風格的故事"},
          {"type":"image_url",
          "image_url":{"url": f"data:image/jpeg;base64,{base64_image_1}"}},
          {"type":"image_url",
          "image_url":{"url": f"data:image/jpeg;base64,{base64_image_2}"}}
        ]
      }
      ],
    "max_tokens": int(num*2.5)
  }
  response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
  dictRes=response.json()
  return dictRes['choices'][0]['message']['content']
```

```python
gr.Interface(
  fn=Vision,
  inputs=[gr.Image(label='1：',sources='upload',type='filepath',width=96,height=96),
      gr.Image(label='2：',sources='upload',type='filepath',width=96,height=96),
      gr.Dropdown(['溫馨','寓言','童話','趣味','懸疑'],label='風格：',value=0),
      gr.Slider(label='字數：',minimum=50,maximum=1000,value=200)],
  outputs=gr.Textbox(label='故事：'),
  title='故事生成器'
).queue().launch()
```
