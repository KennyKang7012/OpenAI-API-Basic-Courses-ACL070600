```python
import base64
import requests
```

```python
def EncodeImage(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
def Vision(image):
  api_key = "OpenAIAPI金鑰"
  base64_image = EncodeImage(image)
  headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
  }
  payload = {
    "model":"gpt-4o",
    "messages":[
      {
        "role":"user",
        "content": [
          {"type":"text","text":"圖像中有甚麼?"},
          {"type":"image_url",
          "image_url":{"url": f"data:image/jpeg;base64,{base64_image}"}
          }
        ]
      }
      ],
    "max_tokens": 300
  }
  response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
  dictRes=response.json() #傳回值轉為字典物件
  return dictRes['choices'][0]['message']['content']
```

```python
gr.Interface(
  fn=Vision,
  inputs=gr.Image(label='來源圖片：',sources='upload',type='filepath',width=256,height=256),
  outputs=gr.Textbox(label='AI描述圖像的內容：')
).queue().launch()
```
