```python
image_url='https://upload.wikimedia.org/wikipedia/commons/thumb/4/42/Artificial-Intelligence.jpg/800px-Artificial-Intelligence.jpg'
def Vision(img_url):
  openai.api_key = 'OpenAIAPI金鑰'
  response = openai.chat.completions.create(
  model="gpt-4o",
  messages=[
    {
      'role':'user',
      'content':[
        {'type':'text','text':'圖像中有甚麼?'},
        {'type':'image_url','image_url':{'url':image_url}}]
    }
    ],
  max_tokens=300)
  return response.choices[0].message.content
```

```python
gr.Interface(
  fn=Vision,
  inputs=gr.Image(label='來源圖片：',value=image_url,height=256,width=256),
  outputs=gr.Textbox(label='AI描述圖像的內容：')
).queue().launch()
```
