```python
import base64
import requests
import io
```

```python
def SaveImage(pil):
  byte_arr = io.BytesIO()
  pil.save(byte_arr,format='PNG')
  return base64.b64encode(byte_arr.getvalue()).decode('utf-8')
```

```python
def Vision(image):
  api_key = "OpenAIAPI金鑰"
  base64_image = SaveImage(image)
  headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
  }
  payload = {
    "model":"gpt-4o",
    "messages":[
      {
        "role":"system",
        "content":"你是一位專業的營養師"
      },
      {

        "role":"user",
        "content": [
          {"type":"text","text":"以少於20字說明圖像中食物，並估計熱量為多少大卡?輸出格式為1.說明：食物說明,2.熱量：0.0 kcal,圖像非食物時輸出:無法估算"},
          {"type":"image_url",
          "image_url":{"url": f"data:image/jpeg;base64,{base64_image}"}
          }
        ]
      }
      ],
    "max_tokens": 300
  }
  response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
  dictRes=response.json()
  return dictRes['choices'][0]['message']['content']
```

```python
gr.Interface(
  fn=Vision,
  inputs=gr.Image(label='食物：',sources='webcam',type='pil',width=256,height=256),
  outputs=gr.Textbox(label='熱量：'),
  title='食物熱量估算'
).queue().launch()
```
