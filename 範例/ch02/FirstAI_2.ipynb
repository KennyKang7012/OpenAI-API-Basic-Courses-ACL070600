```python
import openai
openai.api_key = 'OpenAIAPI金鑰'
```

```python
response = openai.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {'role': 'system', 'content': '你是一位網路行銷小編，請以此身份回覆，請繁體中文回覆'},
        {'role': 'user', 'content': '幫我撰寫人中之龍8的行銷貼文'}
    ]
)
```
