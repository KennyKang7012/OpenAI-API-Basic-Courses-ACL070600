```python
dialogue = []      # 對話記錄串列
```

```python
def get_response(dialogue):
  response = openai.chat.completions.create(
      model = 'gpt-3.5-turbo',
      messages = dialogue,
      stream = True                              # 設定流式傳輸生成回覆
  )
  for chunk in response:                         # response 為可迭代物件
    if chunk.choices[0].delta.content is not None:   # 迭代元素時,防止None文本
      yield chunk.choices[0].delta.content         # 傳出生成回覆文本
```

```python
ask = '花蓮有哪些旅遊景點?'
dialogue.append({'role':'user','content':ask})     # 存放提問資料
for answer in get_response(dialogue):           # 傳入提問呼叫函式生成回覆
  print(answer, end = '')                     # 逐一顯示片段回覆文本
print('\n')
```
