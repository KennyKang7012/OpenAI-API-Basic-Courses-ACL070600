```python
def myGenerator():
  yield 'Mary'                  # 第一批傳回值為字串 'Mary' 字串
  yield 200                     # 第二批傳回值為數值 200
  yield [11,22,33,44]           # 第三批傳回值為串列 [11,22,33,44]
  yield {'role':'user','content':'介紹台北101大樓'}  # 第四批傳回值為字典
```

```python
for gen in myGenerator():     # myGenerator()是生成器物件
  print(gen)
```
