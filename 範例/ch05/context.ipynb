```python
dialogue = []      # 對話記錄串列
def get_response(dialogue):
  response = openai.chat.completions.create(
      model = 'gpt-3.5-turbo',
      messages = dialogue
  )
  return response.choices[0].message.content    # 傳出生成回覆文本
```

```python
saveTalk = 2       # 保留最後2組提問和回覆記錄
def chat(ask):
  global dialogue            # 宣告dialogue屬於全域變數
  dialogue.append({'role':'user','content':ask})          # 存放提問資料
  reply = get_response(dialogue)        # 傳入提問資料呼叫函式生成回覆資料
  dialogue.append({'role':'assistant','content':reply})   # 存放回覆資料
  while len(dialogue) > saveTalk * 2:          # 當提問和回覆記錄超過2組時
    dialogue.pop(0)                          # 把前面的串列元素移除掉
  return reply                          # 傳出回覆文本
```

```python
gr.Interface(
    fn = chat,
    inputs = gr.Textbox(label='輸入您的問題：'),
    outputs= gr.Textbox(label='AI 的回答：'),
    title = 'AI 聊天室'
).queue().launch(share=True)
```
