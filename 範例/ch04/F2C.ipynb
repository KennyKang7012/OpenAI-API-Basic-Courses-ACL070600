```python
def F2c(c_f,degrees):
  if c_f == '攝氏':
    c = degrees
    f = int(9/5*c+32)
    msg = f'攝氏 {c} 度等於華氏 {f} 度'
  else:
    f = degrees
    c = int((f-32)*5/9)
    msg = f'華氏 {f} 度等於攝氏 {c} 度'
  if c > 30:
    img='hot.jpg'
  else:
    img='cold.jpg'
  return msg,img
```

```python
gr.Interface(
  fn=F2c,
  inputs=[
    gr.Radio(choices=['攝氏','華氏'],value='攝氏',label='溫度單位'),
    gr.Slider(minimum=-100,maximum=100,value=0,label='溫度')],
  outputs=[
    gr.Textbox(label='換算結果：'),
    gr.Image(value='cold.jpg',height=200,width=200)],
  title='溫度換算',
).queue().launch()
```
