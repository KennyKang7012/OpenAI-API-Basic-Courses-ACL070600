```python
def Change(姓名, 年, 是否為西元年):
  years = int(年)
  msg = ''
  if 是否為西元年 == True:
    years -= 1911
    msg = '民國'
  else:
    years += 1911
    msg = '西元'
  return f'{姓名}你好！{年}年轉為{msg}年：', years
```

```python
iface = gr.Interface(
	      fn=Change,
	      inputs=['text','text','checkbox'],
	      outputs=['label','number'])
```

```python
iface.queue()
iface.launch()
```
