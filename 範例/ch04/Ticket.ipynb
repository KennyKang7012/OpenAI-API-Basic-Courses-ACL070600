```python
def Price(u_name, u_age, student, local):
  price=0
  if u_age<6:
    price=0
  else:
    if local==True:
      price=30
    elif student==True:
      price=50
    elif u_age >17 and u_age<=64:
      price=100
    else:
      price=50
  return f'{u_name}你好，你的票價是{price}元'
```

```python
name=gr.Textbox(label='輸入你的姓名：')
age=gr.Slider(minimum=-1,maximum=120,step=1,value=50,label='輸入你的年齡：')
stu=gr.Checkbox(label='學生',value=False)
loc=gr.Checkbox(label='當地居民',value=False)
gr.Interface(
  fn=Price,
  inputs=[name,age,stu,loc],
  outputs=gr.Label(label='票價：')
).queue().launch()
```
